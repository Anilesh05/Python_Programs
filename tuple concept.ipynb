```python
# 1.Creating different types of tuples

# Empty tuple
my_tuple = ()
print("\tDisplay the tuple :",my_tuple)


# Tuple having integers
my_tuple = (1, 2, 3)
print("\tDisplay the tuple :",my_tuple)


# Tuple with mixed datatypes
my_tuple = (1, "Hello", 3.4)
print("\tDisplay the tuple :",my_tuple)


# Nested tuple
my_tuple = ("mouse", [8, 4, 6], (1, 2, 3))
print("\tDisplay the tuple :",my_tuple)


# Tuple packing
my_tuple = 3, 4.6, "dog"
print("\t",my_tuple)


# Tuple unpacking
a, b, c = my_tuple
print("\t",a) 
print("\t",b) 
print("\t",c)  


# 2.Accessing Tuple Elements

# Indexing
my_tuple = ('p', 'e', 'r', 'm', 'i', 't')
print("\tDisplay the tuple :",my_tuple)
print("\telement of [0] is :", my_tuple[0])   
print("\telement of [5] is :", my_tuple[5])   


# Nested indexing
n_tuple = ("mouse", [8, 4, 6], (1, 2, 3))
print("\tDisplay the tuple :",n_tuple)
print("\tNested element of [0][3] is :", n_tuple[0][3])       
print("\tNested element of [1][1] is :", n_tuple[1][1])      


# Negative indexing
my_tuple = ('p', 'e', 'r', 'm', 'i', 't')
print("\tDisplay the tuple :",my_tuple)
print("\telement of [-1] is :", my_tuple[-1])  
print("\telement of [-6] is :", my_tuple[-6])  


# Slicing
my_tuple = ('p', 'r', 'o', 'g', 'r', 'a', 'm', 'i', 'z')
print("\tDisplay the tuple :",my_tuple)
print("\tSlice [1:5]:", my_tuple[1:5])   
print("\tSlice [:-7]:", my_tuple[:-7])  
print("\tSlice [7:]:", my_tuple[7:])    
print("\tFull slice[:]:", my_tuple[:])    


# 3. Changing a Tuple
my_tuple = (4, 2, 3, [6, 5])
print("\tDisplay the tuple :",my_tuple)

# Changing a mutable element
my_tuple[3][0] = 9  
print("\tChanged tuple:", my_tuple)


# Reassigning a tuple
my_tuple = ('p', 'r', 'o', 'g', 'r', 'a', 'm', 'i', 'z')
print("\tReassigned tuple:", my_tuple) 


# 4. Deleting a Tuple
my_tuple = ('p', 'r', 'o', 'g', 'r', 'a', 'm', 'i', 'z')


# Deleting an entire tuple
del my_tuple
print("\tTuple deleted successfully!")


# 5. Tuple Methods
my_tuple = ('a', 'p', 'p', 'l', 'e',)
print("\tDisplay the tuple :",my_tuple)
print("\tCount of 'p' in tuple:", my_tuple.count('p')) 
print("\tIndex of 'l' in tuple:", my_tuple.index('l'))  
```
