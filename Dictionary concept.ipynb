```python
# 1. Creating Python Dictionary
# empty dictionary
my_dict = {}
print("\tEmpty Dictionary:", my_dict)

# dictionary with integer keys
my_dict = {1: 'apple', 2: 'ball'}
print("\tDictionary with integer keys:", my_dict)

# dictionary with mixed keys
my_dict = {'name': 'python', 1: [2, 4, 3]}
print("\tDictionary with mixed keys:", my_dict)

# using dict()
my_dict = dict({1:'apple', 2:'ball'})
print("\tUsing dict() with key-value pairs:", my_dict)

# from sequence having each item as a pair
my_dict = dict([(1,'apple'), (2,'ball')])
print("\tUsing dict() with sequence of pairs:", my_dict)

# 2. Accessing Elements from Dictionary
# get vs [] for retrieving elements
my_dict = {'name': 'python', 'age': 20}
print("\tDisplay dictionary : ",my_dict)

print("\tAccessing element using key:", my_dict['name'])
print("\tAccessing element using get() method:", my_dict.get('age'))
print("\tAccessing non-existent key using get():", my_dict.get('address'))


# 3. Changing and Adding Dictionary elements
my_dict = {'name': 'python', 'age': 20}
print("\tDisplay dictionary : ",my_dict)
# update value
my_dict['age'] = 21
print("\tUpdated dictionary:", my_dict)

# add item
my_dict['address'] = 'Downtown'
print("\tDictionary after adding a new key-value pair:", my_dict)


# 4. Removing elements from Dictionary
# create a dictionary
squares = {1: 1, 2: 4, 3: 9, 4: 16, 5: 25}
print("\tDisplay new dictionary : ",squares)
print("\tRemoved item with key 4:", squares.pop(4))
print("\tDictionary after removing item:", squares)

print("\tRemoved arbitrary item:", squares.popitem())
print("\tDictionary after removing arbitrary item:", squares)

squares.clear()
print("\tDictionary after clearing all items:", squares)


# 5. Python Dictionary Methods
# Dictionary Methods
marks = {}.fromkeys(['Math', 'English', 'Science'], 0)

print("\tDictionary created with fromkeys method:", marks)

for item in marks.items():
    print("\tKey-Value pair:", item)

print("\tSorted keys:", list(sorted(marks.keys())))
```
