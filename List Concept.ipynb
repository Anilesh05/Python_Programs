```python
#1. Create list

list1 = [1, 2, 3, 4, 5, 2, 4, 1, 6, 3, 4]

print("\n\tDisplay the list1 : ", list1)

#2. Constructor

list2 = list(('bus', 'car', 'van' ))  

print("\n\tDisplay the list2 : ", list2)


#3. Looping through the list (for loop)


for i in list2:
    print('\t',i)


#4. Copy


listcopy=list1.copy()
print('\n\tCreate a copy of list 1 name is listcopy :',listcopy)


#5. sort


list1.sort()

print('\n\tDisplay after sorting in list 1 :',list1)


#6. reverse


listcopy.sort(reverse=True)

print('\n\tDisplay after reverse sorting in list 1 :',listcopy)


#7. insert


list1.insert(0,"Numbers")

print('\n\tDisplay after inserted in list1:',list1)


#8. append


list1.append("End")

print('\n\tDisplay after append in list1: ',list1)


#9. join operation


listjoin = listcopy+list2

print('\n\tDisplay after join in listcopy + list 2 :',listjoin)


#10. extend


list1.extend(list2)

print('\n\tDisplay after expend in list1:',list1)


#11. length of the list


print('\n\tThe length of list 1 is :',len(list1))



#12. delete

print('\n\tDisplay before Deleting in list1:',list1)

del list1[11]

print('\n\tDisplay after Deleting in list1:  ',list1)


#13. pop


print('\n\tDisplay before popped in list1:',list1)

print('\n\tPopped element in list1:',list1.pop(12))

print('\n\tDisplay after popped in list1:  ',list1)


#14. remove

print('\n\tDisplay before removing in list1:',list1)

list1.remove('Numbers')

print('\n\tDisplay after removing in list1:  ',list1)


#15. searching index

indexing=list2.index("car")

print("\n\tDisplay index of 'car' in list2: ",indexing)


#16. slicing

sliced=list1[5:]

print("\n\tElements sliced from 5th element till the end :",end=' ')
print(sliced)


#17. clear

list1.clear()

print('\n\tDisplay after clear : ',list1)
```
