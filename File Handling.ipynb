```python
import os       


# Content to be written to the file
content ='''Hello! Welcome to sample.txt
This file is for testing purposes.
Good Luck! '''


# Create file

f = open("sample.txt", "x")  
f.close()                    




# Check if a file exists

if os.path.exists("sample.txt"):
    print("File exists.")         
else:
    print("File doesn't exist.")  

    
    
    
# Write to file

f = open("sample.txt", 'w')                  
f.write(content)                                
print("Content has been written to the file.") 
f.close()                                       





# Read from file

f = open("sample.txt", 'r')  
print("\nFile content:-")        
print(f.read())                
f.close()                    





# Read Only Parts of the File

f = open("sample.txt", 'r')  
print("\nRead the 10 first characters :- ")
print(f.read(10))                          
f.close()                     




# Read Lines of the File

f = open("sample.txt", 'r')   
print("\nRead one line of the file :- ")
print(f.readline())                           
f.close()                     




# Append to file

f = open("sample.txt", 'a')                     
f.write("\nThis is an appended line.")       
print("Content has been appended to the file.")  
f.close()                                        





# Loop through the file line by line:

f = open("sample.txt", 'r')   
for x in f:
    print(x) 
print()
f.close()                      





# Delete file

if os.path.exists("sample.txt"):
    os.remove("sample.txt")                 
    print("File deleted successfully.")  
else:
    print("File doesn't exist.")           
    




# Create directory

if not os.path.exists("sample_directory"):
    os.makedirs("sample_directory")          
    print("Directory created successfully.")  
else:
    print("Directory already exists.")        




# Delete directory

if os.path.exists("sample_directory"):
    os.rmdir("sample_directory")         
    print("Directory deleted successfully.") 
else:
    print("Directory doesn't exist.")         
```
